```python
import pandas as pd
import numpy as np
```

```python
df_pdsi = pd.read_csv("processed/wildfire_augmented_pdsi.csv")
df_pr = pd.read_csv("processed/wildfire_augmented_pr.csv")
df_rmax = pd.read_csv("processed/wildfire_augmented_rmax.csv")
df_rmin = pd.read_csv("processed/wildfire_augmented_rmin.csv")
df_tmmn = pd.read_csv("processed/wildfire_augmented_tmmn.csv")
df_tmmx = pd.read_csv("processed/wildfire_augmented_tmmx.csv")
df_vs = pd.read_csv("processed/wildfire_augmented_vs.csv")
df_sph = pd.read_csv("processed/wildfire_augmented_sph.csv")
```

```python
df = pd.merge(df_pdsi, df_pr)
df = pd.merge(df, df_rmax)
df = pd.merge(df, df_rmin)
df = pd.merge(df, df_rmax)
df = pd.merge(df, df_tmmn)
df = pd.merge(df, df_tmmx)
df = pd.merge(df, df_vs)
df = pd.merge(df, df_sph)
df
```
