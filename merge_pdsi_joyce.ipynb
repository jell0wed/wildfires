```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
orig_df = pd.read_csv("wildfires.csv")
orig_df
```

```python
merge_datasets = [{
    'filename': 'rmax_by_FOD_ordered.csv',
    'src_col_name': 'PDSI',
    'tar_col_name': 'RMAX'
},
{
    'filename': 'rmin_by_FOD_ordered.csv',
    'src_col_name': 'PDSI',
    'tar_col_name': 'RMIN'
},
{
    'filename': 'tmmn_by_FOD_ordered.csv',
    'src_col_name': 'max_temp',
    'tar_col_name': 'TMMN'
},
{
    'filename': 'tmmx_by_FOD_ordered.csv',
    'src_col_name': 'max_temp',
    'tar_col_name': 'TMMX'
},
{
    'filename': 'vs_by_FOD_ordered.csv',
    'src_col_name': 'PDSI',
    'tar_col_name': 'VS'
}]
```

```python
for _, dataset in enumerate(merge_datasets):
    ds_df = pd.read_csv(dataset['filename'])
    print("opened %s" % (dataset['filename']))
    ds_df = ds_df.rename(columns={dataset['src_col_name']: dataset['tar_col_name']})
    orig_df = pd.merge(orig_df, ds_df[['FOD_ID',dataset['tar_col_name']]], left_on='FOD_ID', right_on='FOD_ID')
    print("merged %s" % (dataset['filename']))
```

```python
orig_df
```

```python
pd.set_option('display.max_columns', None)
pdsi_data = pd.read_csv("processed/wildfire_augmented_pdsi.csv")
pdsi_data = pdsi_data.rename(columns={'pdsi_first': 'PDSI'})
orig_df = pd.merge(orig_df, pdsi_data[['FOD_ID','PDSI']], left_on='FOD_ID', right_on='FOD_ID')
orig_df
```

```python
def compute_date(row, col_name):
    f_year, doy = row['FIRE_YEAR'], row[col_name]
    if np.isnan(doy):
        return np.nan
    cdate = datetime(f_year, 1, 1) + timedelta(days=np.float64(doy))
    return cdate
    
    
orig_df['DISCOVERY_CDATE'] = orig_df.apply(lambda row: compute_date(row, 'DISCOVERY_DOY'), axis=1)
orig_df['CONT_CDATE'] = orig_df.apply(lambda row: compute_date(row, 'CONT_DOY'), axis=1)
```

```python
pd.set_option('display.max_columns', None)
orig_df
```

```python
orig_df = orig_df.drop(['DISCOVERY_DATE', 'DISCOVERY_DOY', 'DISCOVERY_TIME', 'CONT_DATE', 'CONT_DOY', 'CONT_TIME'], axis=1)
orig_df.columns
```

```python
orig_df.to_csv("wildfires-cleaned.csv")
```
